```python
import numpy as np
import pandas as pd
import os
data = pd.read_csv('./genderClassification.csv')
data = data.drop(columns='Unnamed: 0')
print(data)
```

```python
def getAvgSentence(filename):
    lines, blanklines, sentences, words = 0, 0, 0, 0
    filepath = './FinalProjectBooks/' + filename
    file = open(filepath, 'r', encoding="utf8")

    for line in file:
        lines += 1
        sentences += line.count('.') + line.count('!') + line.count('?')
        tempwords = line.split(None)
        words += len(tempwords)
    avgLength = words/sentences
    file.close()
    all_avg_sentences.append(avgLength) #add to list containing all novel averages
    print(filename, ": ", avgLength)
    
```
