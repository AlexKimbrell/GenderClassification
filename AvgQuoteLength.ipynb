```python
def getQuoteStats(filename):
    filepath = './FinalProjectBooks/' + filename
    file = open(filepath, 'r', encoding="utf8")
    text = []
    quote_lengths = []
    quote_word_lengths = []

    while True:
        quote_length = 0
        words = 1 #Counting spaces so add 1. Ex: "Test" should have length=1
        char = file.read(1)
        if not char:
            #print("***", filename, "***")
            break
        #Check for slanted quotes
        if char == "“": 
            char = file.read(1) #advance to next char after first quotation
            while char != "”": 
                quote_length = quote_length + 1 #add to the length of the individual quote
                char = file.read(1)
                if char == " ":
                    words = words + 1 #To count the number of words(spaces)
                if not char:
                    #print("Infinite Quote")
                    break
            quote_lengths.append(quote_length) #add the length of a single quote to list of quote lengths
            quote_word_lengths.append(words)
            
        #Check for normal quotes  
        elif char == "\"": #Check for slanted quotes
            char = file.read(1) #advance to next char after first quotation
            while char != "\"": 
                quote_length = quote_length + 1 #add to the length of the individual quote
                char = file.read(1)
                if char == " ":
                    words = words + 1 #To count the number of words(spaces)
                if not char:
                    #print("Infinite Quote")
                    break
            quote_lengths.append(quote_length) #add the length of a single quote to list of quote lengths
            quote_word_lengths.append(words)


    #Print out number of quotes found
    tot_num_quotes = len(quote_lengths)
    All_Num_Quotes.append(tot_num_quotes) #Add to list with all books
    #print("Number of quotes found: ", tot_num_quotes)

    #Get the average & median # of words in quotes
    tot_num_avg_quotes = statistics.mean(quote_word_lengths)
    All_Num_Avg_Quotes.append(tot_num_avg_quotes) #Add to list with all books
    #print("AVERAGE WORDS PER QUOTE: ", tot_num_avg_quotes)
    #print("MEDIAN WORDS PER QUOTE: ", statistics.median(quote_word_lengths))
    #print("MAX: ", max(quote_word_lengths))
    #print("MIN: ", min(quote_word_lengths))
    #print()
    #print()


    #Close File I/O Stream
    file.close()
```

```python
#Get the adjectives from a book
def getAdj(filename):
    filepath = './FinalProjectBooks/' + filename
    file = open(filepath, 'r', encoding="utf8")
    count = 0
    for sentence in file:
        text = nltk.word_tokenize(sentence)
        parts_of_speech = nltk.pos_tag(text)
        for val in parts_of_speech:
            if(val[1] == "JJ"): #If word is an adjective, increment count
                count = count + 1
    print("# of Adjectives: ",count)
    All_Num_Adj.append(count) #Add to list containing each book
    file.close()
```
