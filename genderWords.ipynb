```python
#Alex Kimbrell
#CSCI 4360 Final Project
#All books should be *.txt and stored in './FinalProjectBooks/'
import numpy as np
import pandas as pd
import statistics
import os
import nltk
data = pd.read_csv('./genderClassification.csv')
data = data.drop(columns='Unnamed: 0')
print(data)
```

```python
#Determine if certain words appear more in female/male novels
male_words = ["chief", "rear", "civil", "big", "absolute", "enemy", "fellow", "king", "public", "contact"]
female_words = ["pillows", "lace", "curls", "dress", "china", "skirt", "curtains", "cups", "sheets", "shrugged"]
```

```python
#Get the adjectives from a book
def genderWords(filename):
    filepath = './FinalProjectBooks/' + filename
    file = open(filepath, 'r', encoding="utf8")
    male_count = 0
    female_count = 0
    for sentence in file:
        text = nltk.word_tokenize(sentence)
        parts_of_speech = nltk.pos_tag(text)
        for val in parts_of_speech:
            if val[0].lower() in male_words: #Masculine Word is found
                male_count = male_count+1
            elif val[0].lower() in female_words: #Feminin Word is found
                female_count = female_count+1
                
    print("# of Feminine Words: ",female_count, "   # of Masculine Words: ",male_count)
    male_word_counts.append(male_count)
    female_word_counts.append(female_count)
    file.close()
```

```python
#Get Female/Male Count for all novels
male_word_counts = []
female_word_counts = []

count = 1
for file in data['Novel']:
    print(count,"/",215, " (", file, ")")
    genderWords(file+".txt")
    count = count + 1
print("Gender Word Stats: DONE")
```

```python
#Add male word count to dataframe
data['Masculine Words'] = male_word_counts

#Add female word count to dataframe
data['Feminine Words'] = female_word_counts

#Save dataframe to csv
data.to_csv('./genderClassificationPlus.csv')
```
