```python
#Alex Kimbrell
#CSCI 4360 Final Project
#All books should be *.txt and stored in './FinalProjectBooks/'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('./genderClassificationPlus.csv')
data = data.drop(columns='Unnamed: 0')
```

```python
#Dimensionality Reduction
from sklearn.decomposition import IncrementalPCA
X = data.drop(columns=['Novel', 'Male/Female'])
ipca = IncrementalPCA(n_components=2, batch_size=5)
ipca.fit(X)
IncrementalPCA(batch_size=3, copy=True, n_components=2, whiten=False)
newData = ipca.transform(X)
```

```python
#Plot the results
dfcolor = pd.DataFrame([['Male', 'red'],['Female', 'blue']], columns=['Male/Female', 'Color'])
mergeddf = pd.merge(data, dfcolor)

male_data = mergeddf[mergeddf['Color'] == 'red']
female_data = mergeddf[mergeddf['Color'] == 'blue']
plt.scatter(newData[:188,0], newData[:188,1], color=(male_data['Color']), label="Male")
plt.scatter(newData[188:,0], newData[188:,1], color=(female_data['Color']), label="Female")

plt.legend()
plt.title('Gender Classification')
```
